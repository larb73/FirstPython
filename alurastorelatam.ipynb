### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
# prompt: Sumando los valores de la columna Precio en cada conjunto de datos

# Suma de la columna 'Precio' para cada tienda
suma_tienda1 = tienda['Precio'].sum()
suma_tienda2 = tienda2['Precio'].sum()
suma_tienda3 = tienda3['Precio'].sum()
suma_tienda4 = tienda4['Precio'].sum()

print(f"Total Ventas Tienda 1: ${suma_tienda1}")
print(f"Total Ventas Tienda 2: ${suma_tienda2}")
print(f"Total Ventas Tienda 3: ${suma_tienda3}")
print(f"Total Ventas Tienda 4: ${suma_tienda4}")
```

```python
# prompt: Suma los valores de la columna precio en cada conjunto de datos y muestra en grafico por tienda

import matplotlib.pyplot as plt

# Datos para el gráfico
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
sumas = [suma_tienda1, suma_tienda2, suma_tienda3, suma_tienda4]

# Crear el gráfico de barras
plt.figure(figsize=(8, 6))  # Ajustar el tamaño de la figura
plt.bar(tiendas, sumas, color=['skyblue', 'lightcoral', 'lightgreen', 'gold']) #Colores a las barras
plt.xlabel('Tiendas')
plt.ylabel('Total Facturacion')
plt.title('Total Facturación por Tienda')
plt.grid(axis='y', linestyle='--', alpha=0.7) #Agregar una cuadrícula al gráfico
plt.show()
```

# 4. Productos más y menos vendidos

```python
# prompt: cuenta cada producto y suma el mismo producto en cada tienda, muestra la mayor y menor suma de producto por cada tienda y genera adicional un grafico de barras por cada tienda

import pandas as pd
import matplotlib.pyplot as plt

# ... (previous code)

def analyze_products(df, tienda_name):
    product_counts = df.groupby('Producto')['Precio'].sum()
    max_product = product_counts.idxmax()
    min_product = product_counts.idxmin()
    max_sum = product_counts.max()
    min_sum = product_counts.min()

    print(f"\nAnálisis de productos para {tienda_name}:")
    print(f"Producto más vendido: {max_product}, Cant. Maxima: {max_sum}")
    print(f"Producto menos vendido: {min_product}, Cant. Minima: {min_sum}")

    product_counts.plot(kind='bar', figsize=(10, 6))
    plt.title(f"precios por producto en {tienda_name}")
    plt.xlabel("Producto")
    plt.ylabel("precios")
    plt.xticks(rotation=45, ha='right')  # Rotar etiquetas del eje x
    plt.tight_layout() # Ajustar diseño
    plt.show()

analyze_products(tienda, "Tienda 1")
analyze_products(tienda2, "Tienda 2")
analyze_products(tienda3, "Tienda 3")
analyze_products(tienda4, "Tienda 4")
```
